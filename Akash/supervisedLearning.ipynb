```python
import matplotlib as plt
import seaborn as sns
import matplotlib.patches as mpatches
# import numpy as np

#Plot
sns.countplot(x='target', data=df, palette=['skyblue','orange'])

healthy = mpatches.Patch(color='skyblue', label='healthy')
unhealthy = mpatches.Patch(color='orange', label='unhealthy')

plt.legend(handles=[healthy, unhealthy],title='Conditions',fontsize=12,handlelength=2,handleheight=2)
plt.margins(x=0.2)
plt.margins(y=0.7)

plt.show()
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

# Rename columns
df.columns = ["age", "sex", "cp", "trestbps", "chol", "fbs", "restecg", 
              "thalach", "exang", "oldpeak", "slope", "ca", "thal", "target"]

# Horizontal boxplot
df.boxplot(column=list(df.columns), showmeans=True, vert=False)
plt.show()
```

```python
#scattern matrix

scatter_matrix(df.drop(columns=["target"]),figsize=(15,15))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

corrMatrix = df.corr()
fig,ax=plt.subplots(figsize=(15,15))
sns.heatmap(corrMatrix,annot=True , linewidths=0.5, cmap="coolwarm", ax=ax)
plt.show()
```
